# 1.One Hot Encoding & Dummy Variables

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as ptl
import seaborn as sns
```

```python
#import Dataset from Github 

ds1 = sns.load_dataset("tips")
ds1
```

Convert all the Categorical variable into number form By creating extra column Like...For sex we crete 2 columns(Male and female) and if male is present in a row then put '1' in the male colunm and '0' in female column.

##### This process is known as One Hot Encoding. And filling the new columns with variables are called Dummy Variables.

```python
#get_dummies() function creates the colunm and fill the dummy variables
dummy_var = pd.get_dummies(ds1)
dummy_var
```

```python
#Here we have some extra columns like male and female for sex.   ....we can get the result with one single colunm. so we remove one column .

dummy_var2  =  pd.get_dummies(ds1, drop_first=True)
dummy_var2
```

# 2.One Hot Encoding using Scikit-learn

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oh_enc = OneHotEncoder(sparse=False , drop='first')
oh_enc
```

```python
oh_enc_arr = oh_enc.fit_transform(ds1[['sex' , 'smoker' , 'day' , 'time']])
oh_enc_arr
```

```python
dummy_var.keys()
```

```python
oh_enc_df = pd.DataFrame(oh_enc_arr, columns=['sex_Female',
       'smoker_No', 'day_Fri', 'day_Sat', 'day_Sun',
       'time_Dinner'])
oh_enc_df
```
