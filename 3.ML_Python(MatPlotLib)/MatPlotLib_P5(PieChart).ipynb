# Pie Chart

```python
import matplotlib.pyplot as plt
```

```python
classes = ["python" , "R" , "ML" , "AI" , "DataScience"]
classes1 = [45,15,35,25,30]

plt.pie([1])
plt.show()
```

```python
classes = ["python" , "R" , "ML" , "AI" , "DataScience"]
classes1 = [45,15,35,25,30]

plt.pie(classes1)
plt.show()
```

```python
classes = ["python" , "R" , "ML" , "AI" , "DataScience"]
classes1 = [45,15,35,25,30]

plt.pie(classes1 , labels=classes)
plt.show()
```

```python
classes = ["python" , "R" , "ML" , "AI" , "DataScience"]
classes1 = [45,15,35,25,30]
explode = [0,0.1,0,0,0]

plt.pie(classes1 , labels=classes , explode=explode)
plt.show()
```

```python
classes = ["python" , "R" , "ML" , "AI" , "DataScience"]
classes1 = [45,15,35,25,30]
explode = [0,0.1,0,0,0]

plt.figure(figsize=(12,5))
plt.pie(classes1 , labels=classes , explode=explode , autopct="%0.2f%%")
plt.show()
```

```python
classes = ["python" , "R" , "ML" , "AI" , "DataScience"]
classes1 = [45,15,35,25,30]
explode = [0,0.1,0,0,0]

plt.figure(figsize=(16,9))
plt.pie(classes1 , labels=classes , explode=explode , autopct="%0.2f%%")

plt.legend()
plt.show()
```
