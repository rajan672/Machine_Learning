# Histogram Plot

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
stu_age1 = np.random.randint(18,45 ,(100))
print(stu_age1)

stu_age2 = np.random.randint(15,40, (100))
print(stu_age2)
```

```python
#hist() function for histogram
plt.hist(stu_age1)

plt.title("Students age" , fontsize = 20)
plt.xlabel("Students Age" , fontsize=20)
plt.ylabel("No. of students" , fontsize=20)
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(stu_age1 , bins)

plt.title("Students age" , fontsize = 20)
plt.xlabel("Students Age" , fontsize=20)
plt.ylabel("No. of students" , fontsize=20)
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(stu_age1 , bins , rwidth = 0.8)

plt.title("Students age" , fontsize = 20)
plt.xlabel("Students Age" , fontsize=20)
plt.ylabel("No. of students" , fontsize=20)
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(stu_age1 , bins , rwidth = 0.8 , orientation = 'horizontal')

plt.title("Students age" , fontsize = 20)
plt.xlabel("Students Age" , fontsize=20)
plt.ylabel("No. of students" , fontsize=20)
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(stu_age1 , bins , rwidth = 0.8 , orientation = 'horizontal' , color = "g")

plt.title("Students age" , fontsize = 20)
plt.xlabel("Students Age" , fontsize=20)
plt.ylabel("No. of students" , fontsize=20)
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(stu_age1 , bins , rwidth = 0.8 , color = "g")
plt.hist(stu_age2 , bins , rwidth = 0.8 , color = "r")
plt.title("Students age" , fontsize = 20)
plt.xlabel("Students Age" , fontsize=20)
plt.ylabel("No. of students" , fontsize=20)
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist([stu_age1 , stu_age2], bins , rwidth = 0.8 , color = ["g" , "r" ])

plt.title("Students age" , fontsize = 20)
plt.xlabel("Students Age" , fontsize=20)
plt.ylabel("No. of students" , fontsize=20)
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.figure(figsize = (16,9))
plt.hist([stu_age1 , stu_age2], bins , rwidth = 0.8 , color = ["g" , "r" ])

plt.title("Students age" , fontsize = 20)
plt.xlabel("Students Age" , fontsize=20)
plt.ylabel("No. of students" , fontsize=20)
plt.show()
```
