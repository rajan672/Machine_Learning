# Line Plot

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
Day = np.arange(1,16)
print (Day)
```

```python
Temperature = [36.6 , 37 , 37.7 , 39 , 40.1 , 43 , 43.3 , 45 , 45.6 , 40.1 , 37 , 34 , 33 , 48 , 49]
print (Temperature)
```

```python
#The plot() function in pyplot module of matplotlib library is used to make a 2D hexagonal binning plot of points x, y.
plt.plot(Day , Temperature)
plt.show()
```

```python
#Plotting the points
plt.plot(Day , Temperature)

#Title of the Plot
plt.title("Delhi Temperature ")

#X- axis
plt.xlabel("Day")
#Y-axis
plt.ylabel("Temperature")

#Display the Plot
plt.show()
```

```python
#Plotting the points
plt.plot(Day , Temperature)

# Give the plot starting and ending points   axis([xmin, xmax, ymin, ymax])
plt.axis([0,30,0,50])

#Title of the Plot
plt.title("Delhi Temperature ")

#X- axis
plt.xlabel("Day")
#Y-axis
plt.ylabel("Temperature")

#Display the Plot
plt.show()
```

# Line Plot Part-2

```python
# plot( X , Y , Color , Marker)
plt.plot(Day , Temperature , color = "Red", marker = "o") #  'o' = Circle Marker
plt.title("Delhi Temperature ")
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.show()
```

```python
# plot( X , Y , Color , Marker  , linestyle)
plt.plot(Day , Temperature , color = "Red", marker = "o" , linestyle = "--") #  'o' = Circle Marker
plt.title("Delhi Temperature ")
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.show()
```

```python
#Minimize the plot function
plt.plot(Day , Temperature , "ro--") #  'o' = Circle Marker
plt.title("Delhi Temperature ")
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.show()
```

```python
plt.plot(Day , Temperature , "ro--") #  'o' = Circle Marker
plt.title("Delhi Temperature " , fontsize = 20)
plt.xlabel("Day" , fontsize = 20)
plt.ylabel("Temperature" , fontsize = 20)
plt.show()
```

```python
plt.plot(Day , Temperature , "ro--") #  'o' = Circle Marker
plt.title("Delhi Temperature " , fontsize = 20)
plt.xlabel("Day" , fontsize = 20)
plt.ylabel("Temperature" , fontsize = 20)
#lezend()
plt.legend("[Line]" , loc = 4)
plt.show()
```
