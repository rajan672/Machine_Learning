```python
df = pd.read_csv("MagicBricks.csv" , nrows=1000)
df
```

```python
df.shape
```

```python
X = df["Price"]
Y = df["BHK"]
```

```python
plt.figure(figsize=(25,9))
plt.scatter(X,Y)
```

```python
plt.figure(figsize=(25,9))
plt.scatter(X,Y)

plt.title("Housing price")
plt.xlabel("Price")
plt.ylabel("BHK")
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.scatter(X,Y , c='r')

plt.title("Housing price")
plt.xlabel("Price")
plt.ylabel("BHK")
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.scatter(X,Y , c='r')
plt.scatter(X,df["Bathroom"] , c='g')
plt.title("Housing price")
plt.xlabel("Price")
plt.ylabel("BHK")
plt.show()
```
