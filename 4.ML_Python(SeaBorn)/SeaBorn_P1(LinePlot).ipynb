# Line Plot in Sea Born

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Day1 =np.arange(1,16)
Temperature1 = [36.6 , 37 , 37.7 , 39 , 40.1 , 43 , 43.3 , 45 , 45.6 , 40.1 , 37 , 34 , 33 , 48 , 49]

df1 = pd.DataFrame({"Days": Day1 , "Temperature" :Temperature1}) # DataFrame of Pandas

sns.lineplot(x = "Days" ,y = "Temperature" , data= df1) # seaborn for lineplot

plt.show() 
```

```python
#import Dataset from Github 

ds1 = sns.load_dataset("tips")
ds1
```

```python
sns.lineplot(x="total_bill" , y="tip" , data=ds1)
plt.show()
```

```python
sns.lineplot(x="total_bill" , y="size" , data=ds1)
plt.show()
```

```python
sns.lineplot(x="total_bill" , y="size" , data=ds1 , hue = "sex")
plt.show()
```

```python
sns.lineplot(x="total_bill" , y="size" , data=ds1 , hue = "sex")

plt.title("Line Plot " , fontsize = 20)
plt.xlabel("total_bill" , fontsize = 20)
plt.ylabel("size" , fontsize = 20)
plt.show()
```

```python
sns.lineplot(x="size" ,y="total_bill" , data=ds1 , hue = "sex" ,style= "sex")
```
